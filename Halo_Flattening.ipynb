```python
import matplotlib.pyplot as plt
from snap_tools.plotting import plot_tools
import snap_tools.io as io
import snap_tools.manipulate as man
import numpy as np
folder = '/usr/users/spardy/coors/gadget/data/output_dehnen/'
%matplotlib inline
```

```python
fig, ax = plt.subplots(4,2,figsize=(10,10))

ax[0]



for j,i in enumerate(xrange(0,40,5)):
    snap = io.load_snap(folder+'snap_'+str(i).zfill(3))
    pos = snap.pos['halo']
    x = pos[:, 0]
    y = pos[:, 1]
    z = pos[:, 2]

    r = np.sqrt(x**2+y**2)

    r = np.log10(r)
    density, binEdges = np.histogram(r, bins=512, range=[-1, 2])
    binEdges = 10**binEdges
    radius = 0.5*(binEdges[1:]+binEdges[:-1])
    vol = 4./3.*np.pi*(binEdges[1:]**3-binEdges[:-1]**3)
    density = density/vol
    ax[j // 2][j % 2].plot(radius, density, label='R')

    r = np.log10(np.sqrt(z**2))
    density, binEdges = np.histogram(r, bins=512, range=[-1, 2])
    binEdges = 10**binEdges
    radius = 0.5*(binEdges[1:]+binEdges[:-1])
    vol = 4./3.*np.pi*(binEdges[1:]**3-binEdges[:-1]**3)
    density = density/vol
    ax[j // 2][j % 2].plot(radius, density, label='Z')
    ax[j // 2][j % 2].set_yscale('log')
    ax[j // 2][j % 2].set_xscale('log')

```
